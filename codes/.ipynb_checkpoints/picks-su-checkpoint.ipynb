```python
%matplotlib widget

import obspy
import matplotlib.pyplot as plt
import numpy as np
import bruges
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

nome_do_arquivo_su = 'plots-files/6.su'
posicao_da_fonte = -49

aplicar_filtro_fk = True
fk_slope_min = 0.000005
fk_slope_max = 999

aplicar_filtro_banda = False
f1, f2, f3, f4 = 20, 30, 130, 150

sobrepor_picks_antigos = False
arquivo_picks_antigos = 'picks_10su.txt'

ganho_visual = 2.0
clipping_percentil = 98
usar_preenchimento = True
plotar_por_indice = False
NG = 96
dx = 1.0
g1 = 0.0

print(f"--- Carregando o arquivo: {nome_do_arquivo_su} ---")
stream = obspy.read(nome_do_arquivo_su, format='SU')
print("Arquivo SU carregado com sucesso!")

if aplicar_filtro_banda:
    stream.filter('bandpass', freqmin=f2, freqmax=f3, corners=4, zerophase=True)
    print(f"Filtro Passa-Banda aplicado entre {f2} Hz e {f3} Hz.")

pos_geofones = g1 + np.arange(NG) * dx
for i, tr in enumerate(stream):
    pos_geofone = pos_geofones[i]
    tr.stats.offset = abs(pos_geofone - posicao_da_fonte)
    tr.stats.geophone_position = pos_geofone
stream.sort(keys=['geophone_position'])

if aplicar_filtro_fk:
    print(f"\nAplicando filtro F-K para rejeitar eventos com velocidade < {1/fk_slope_min:.0f} m/s...")
    try:
        data_array = np.array([tr.data for tr in stream]).T
        dt = stream[0].stats.delta
        filtered_data = bruges.filters.fk_filter(data_array,
                                                 dt=dt,
                                                 dx=dx,
                                                 s_min=fk_slope_min,
                                                 s_max=fk_slope_max,
                                                 kind='rejection')
        for i, tr in enumerate(stream):
            tr.data = filtered_data[:, i]
        print("Filtro F-K aplicado com sucesso.")
    except Exception as e:
        print(f"ERRO ao aplicar o filtro F-K: {e}")

print("\nGeometria atribuída e traços ordenados por POSIÇÃO DO GEOFONE.")

print("\n>>> GRÁFICO INTERATIVO ABAIXO <<<")
picks = [] 
pick_plots = []
fig, ax = plt.subplots(figsize=(12, 8))
if stream:
    all_data_abs = np.abs(np.concatenate([tr.data for tr in stream]))
    clip_val = np.percentile(all_data_abs, clipping_percentil)
    if clip_val == 0: clip_val = 1
else:
    clip_val = 1
all_geophone_positions = np.array([tr.stats.geophone_position for tr in stream])
for i, tr in enumerate(stream):
    x_position = i if plotar_por_indice else tr.stats.geophone_position
    trace_clip_val = np.percentile(np.abs(tr.data), clipping_percentil)
    if trace_clip_val == 0: trace_clip_val = 1
    data_normalized = tr.data / trace_clip_val
    excursion = data_normalized * (dx * ganho_visual)
    if usar_preenchimento:
        positive_excursion = np.where(excursion > 0, excursion, 0)
        ax.fill_betweenx(tr.times(), x_position, x_position + positive_excursion, facecolor='black', linewidth=0)
        ax.plot(x_position + excursion, tr.times(), 'k-', linewidth=0.5, label='_nolegend_')
    else:
        ax.plot(x_position + excursion, tr.times(), 'k-', linewidth=0.8, label='_nolegend_')
if sobrepor_picks_antigos:
    try:
        tempos_antigos, posicoes_antigas = np.loadtxt(arquivo_picks_antigos, unpack=True)
        ax.plot(posicoes_antigas, tempos_antigos, '+', color='cyan', markersize=10, mew=2, label=f'Picks Antigos ({arquivo_picks_antigos})')
        print(f"Arquivo de picks '{arquivo_picks_antigos}' sobreposto no gráfico.")
    except Exception as e:
        print(f"AVISO: Não foi possível carregar ou plotar o arquivo de picks antigos: {e}")
def on_key(event):
    if event.key == 'a':
        ix, iy = event.xdata, event.ydata
        if ix is not None and iy is not None:
            if plotar_por_indice:
                idx_mais_proximo = int(round(ix))
                posicao_correta = all_geophone_positions[idx_mais_proximo]
            else:
                idx_mais_proximo = np.argmin(np.abs(all_geophone_positions - ix))
                posicao_correta = all_geophone_positions[idx_mais_proximo]
            picks.append({'pos': posicao_correta, 't': iy})
            x_marker_pos = idx_mais_proximo if plotar_por_indice else posicao_correta
            plot_obj = ax.plot(x_marker_pos, iy, 'rx', markersize=8, markeredgewidth=2, label='_nolegend_')
            pick_plots.append(plot_obj[0])
            print(f"Pick ADICIONADO: Posição={posicao_correta:.2f} m, Tempo={iy:.4f} s")
            plt.gcf().canvas.draw_idle()
    elif event.key == 'd':
        if picks:
            picks.pop(); last_plot = pick_plots.pop(); last_plot.remove()
            print("Último pick REMOVIDO.")
            plt.gcf().canvas.draw_idle()
fig.canvas.mpl_connect('key_press_event', on_key)
if plotar_por_indice:
    ax.set_xlabel("Índice do Traço")
    ax.set_xlim(-1, len(stream))
else:
    ax.set_xlabel("Posição do Geofone (m)")
    ax.set_xlim(g1 - dx, g1 + (NG-1)*dx + dx)
ax.set_title(f"Picking Interativo - Arquivo: {nome_do_arquivo_su} (Fonte em {posicao_da_fonte}m)")
ax.set_ylabel("Tempo (s)")
ax.invert_yaxis()
ax.legend()
plt.grid(True, linestyle='--' , alpha=0.6)
plt.tight_layout()
plt.savefig('n6-grafico')
plt.show()
```

```python
if picks:
    picks.sort(key=lambda p: p['pos'])
    dados_para_salvar = np.array([(p['t'], p['pos']) for p in picks])
    nome_arquivo_final = f'picks_{nome_do_arquivo_su.replace("/", "-")}.txt'
    np.savetxt(nome_arquivo_final, dados_para_salvar, fmt='%.6f %.6f')
    print(f"\nSUCESSO! Arquivo salvo em: '{nome_arquivo_final}'")
    print("Formato: Coluna 1 = Tempo(s), Coluna 2 = Posição(m)")
else:
    print("\nNenhum pick foi feito. Faça os picks no gráfico acima e execute esta célula novamente.")
```
