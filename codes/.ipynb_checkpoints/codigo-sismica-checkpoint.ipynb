```python
import numpy as np
import matplotlib.pyplot as plt

analises = {
    'n1': {
        'arquivo': 'picks/n1m2',
        'cor_dados': 'deepskyblue',
        'segmentos': {
            'Refratada': {'dist_min': 0, 'dist_max': 99999, 'cor': 'deepskyblue'},
            'Reta 2': {'dist_min': 0, 'dist_max': 0, 'cor': 'darkblue'},
        }
    },
    'n4': {
        'arquivo': 'picks/n4',
        'cor_dados': 'darkblue',
        'segmentos': {
            'Direta': {'dist_min': 83, 'dist_max': 9999, 'cor': 'darkblue'},
            'Refratada': {'dist_min': 0, 'dist_max': 70, 'cor': 'darkblue'},
        }
    },
    'n6direita': {
        'arquivo': 'picks/n6direita',
        'cor_dados': 'darkgreen',
        'segmentos': {
            'Refratada': {'dist_min': 0, 'dist_max': 40, 'cor': 'darkgreen'},
            'Refratada': {'dist_min': 75, 'dist_max': 999, 'cor': 'lime'},
        }
    },
    'n6esquerda': {
        'arquivo': 'picks/n6esquerda',
        'cor_dados': 'lime',
        'segmentos': {
            'Refratada': {'dist_min': 0, 'dist_max': 45, 'cor': 'green'},
            'Reta 2': {'dist_min': 45, 'dist_max': 999, 'cor': 'darkgreen'},
        }
    },
    'n13': {
        'arquivo': 'picks/n13',
        'cor_dados': 'darkred',
        'segmentos': {
            'Direta': {'dist_min': 0, 'dist_max': 20, 'cor': 'darkred'},
            'Refratada': {'dist_min': 19, 'dist_max': 70, 'cor': 'darkred'},
        }
    },
    'n13-curva': {
        'arquivo': 'picks/n13',
        'cor_dados': 'darkred',
        'segmentos': {
            'Direta': {'dist_min': 0, 'dist_max': 0, 'cor': 'darkred'},
            'Refratada': {'dist_min': 83, 'dist_max': 999999, 'cor': 'darkred'},
        }
    },
    'n12': {
        'arquivo': 'picks/n12n2',
        'cor_dados': 'indianred',
        'segmentos': {
             'Refratada': {'dist_min': 0, 'dist_max': 99999, 'cor': 'indianred'},
             'Reta 2': {'dist_min': 0, 'dist_max': 0, 'cor': 'firebrick'},
        }
    },
    'n2': {
        'arquivo': 'picks/n2n2',
        'cor_dados': 'blue',
        'segmentos': {
            'Refratada': {'dist_min': 0, 'dist_max': 999999, 'cor': 'blue'},
            'Reta 2': {'dist_min': 0, 'dist_max': 0, 'cor': 'darkblue'},
        }
    },
    'n10': {
        'arquivo': 'picks/n10n',
        'cor_dados': 'firebrick',
        'segmentos': {
            'Refratada': {'dist_min': 18, 'dist_max': 999999, 'cor': 'firebrick'},
            'Reta 2': {'dist_min': 0, 'dist_max': 0, 'cor': 'firebrick'},
        }
    },
}

print("--- Iniciando Processamento de Dados ---")
dados_processados = {}

for nome_tiro, config in analises.items():
    nome_arquivo = config['arquivo']
    print(f"\n--- Processando o '{nome_tiro}': {nome_arquivo} ---")
    
    try:
        tempos, posicoes = np.loadtxt(nome_arquivo, unpack=True)
    except Exception as e:
        print(f"ERRO ao carregar o arquivo '{nome_arquivo}': {e}")
        continue

    dados_processados[nome_tiro] = {
        'posicoes': posicoes,
        'tempos': tempos,
        'cor_dados': config['cor_dados'],
        'ajustes': []
    }

    if 'segmentos' in config:
        for nome_segmento, limites in config['segmentos'].items():
            mascara = (posicoes >= limites['dist_min']) & (posicoes <= limites['dist_max'])
            pos_segmento = posicoes[mascara]
            t_segmento = tempos[mascara]
            
            if len(pos_segmento) < 2:
                continue
            
            inclinacao, intercepto = np.polyfit(pos_segmento, t_segmento, 1)
            velocidade = abs(1 / inclinacao)
            
            dados_processados[nome_tiro]['ajustes'].append({
                'x': pos_segmento,
                'y': inclinacao * pos_segmento + intercepto,
                'cor': limites['cor'],
                'nome_segmento': nome_segmento,
                'velocidade': velocidade,
            })
            print(f"  -> Segmento '{nome_segmento}' calculado com velocidade {velocidade:.0f} m/s")

print("\n--- Gerando Figura 1: Apenas os Pontos de Pick ---")
fig1, ax1 = plt.subplots(figsize=(12, 8))
for nome_tiro, dados in dados_processados.items():
    ax1.scatter(dados['posicoes'], dados['tempos'], label=nome_tiro, color=dados['cor_dados'], alpha=0.8, s=50)
ax1.set_xlabel('Posição do Receptor (m)')
ax1.set_ylabel('Tempo (s)')
ax1.set_title('Análise Comparativa de Picks (Pontos)')
ax1.grid(True)
ax1.invert_yaxis()
ax1.legend(loc='upper left', bbox_to_anchor=(1.02, 1), title="Tiros")
plt.tight_layout(rect=[0, 0, 0.85, 1])
nome_figura1 = 'analise_comparativa_picks.png'
plt.savefig(nome_figura1, dpi=300, bbox_inches='tight')
print(f"Gráfico de pontos salvo como: {nome_figura1}")
plt.show()

print("\n--- Gerando Figura 2: Apenas as Retas de Ajuste ---")
fig2, ax2 = plt.subplots(figsize=(12, 8))

legend_y_pos = 1.0
line_height = 0.05

for nome_tiro, dados in dados_processados.items():
    for ajuste in dados['ajustes']:
        ax2.plot(ajuste['x'], ajuste['y'], '-', color=ajuste['cor'], lw=2.5)

    ax2.text(1.02, legend_y_pos, f"--- {nome_tiro} ---", transform=ax2.transAxes, fontsize=10, weight='bold', verticalalignment='top')
    legend_y_pos -= line_height

    for ajuste in dados['ajustes']:
        legenda_texto = f"■ {ajuste['nome_segmento']}: {ajuste['velocidade']:.0f} m/s"
        ax2.text(1.02, legend_y_pos, legenda_texto,
                 transform=ax2.transAxes,
                 fontsize=10,
                 color=ajuste['cor'],
                 verticalalignment='top')
        legend_y_pos -= line_height

    ax2.text(1.02, legend_y_pos, "-" * 25, transform=ax2.transAxes, fontsize=10, verticalalignment='top')
    legend_y_pos -= line_height

ax2.set_xlabel('Posição do Receptor (m)')
ax2.set_ylabel('Tempo (s)')
ax2.set_title('Análise Comparativa (Retas de Ajuste)')
ax2.grid(True)
ax2.invert_yaxis()

plt.tight_layout(rect=[0, 0, 0.80, 1])

nome_figura2 = 'analise_retas_ajustadas.png'
plt.savefig(nome_figura2, dpi=300, bbox_inches='tight')
print(f"Gráfico de retas salvo como: {nome_figura2}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================================================================
# PAINEL DE CONTROLE - CONFIGURAÇÕES DO USUÁRIO
# ==============================================================================

arquivo_direto = 'picks/n13'
segmentos_direto = {
    "Onda Direta": { "dist_min": 0, "dist_max": 20, "cor": "red" },
    "Refratada 1": { "dist_min": 83, "dist_max": 999999, "cor": "darkred" },
}

arquivo_reverso = 'picks/n4'
segmentos_reverso = {
    "Onda Direta": { "dist_min": 83, "dist_max": 99999, "cor": "blue" },
    "Refratada 1": { "dist_min": 0, "dist_max": 70, "cor": "darkblue" },
}

distancia_total_AB = 100.0

corte_reciproco_min_m = 19.0
corte_reciproco_max_m = 85.0

eixo_x_min = 15.0
eixo_x_max = 90.0
profundidade_min_plot = -8.0
profundidade_max_plot = -5.0

def processar_tiro(nome_do_arquivo, segmentos_para_analise):
    print(f"\n--- Processando o arquivo: {nome_do_arquivo} ---")
    try:
        tempos_brutos, posicoes_brutas = np.loadtxt(nome_do_arquivo, unpack=True)
    except Exception as e:
        print(f"ERRO ao carregar o arquivo: {e}")
        return None
    indices_sorted = np.argsort(posicoes_brutas)
    posicoes = posicoes_brutas[indices_sorted]
    tempos = tempos_brutos[indices_sorted]
    print(f"Arquivo carregado com {len(posicoes)} pontos.")
    resultados = {'posicoes': posicoes, 'tempos': tempos, 'velocidades': {}, 'interceptos': {}, 'ajustes': {}}
    for nome, limites in segmentos_para_analise.items():
        mascara = (posicoes >= limites['dist_min']) & (posicoes <= limites['dist_max'])
        pos_segmento, t_segmento = posicoes[mascara], tempos[mascara]
        if len(pos_segmento) < 2:
            print(f"AVISO: Segmento '{nome}' tem pontos insuficientes.")
            continue
        inclinacao, intercepto = np.polyfit(pos_segmento, t_segmento, 1)
        velocidade = abs(1 / inclinacao)
        resultados['velocidades'][nome] = velocidade
        resultados['interceptos'][nome] = intercepto
        resultados['ajustes'][nome] = {'x': pos_segmento, 'y': inclinacao * pos_segmento + intercepto, 'cor': limites['cor'], 'label': f'Ajuste {nome} ({velocidade:.0f} m/s)'}
        print(f"  -> Segmento '{nome}': Velocidade = {velocidade:.2f} m/s")
    return resultados

dados_direto = processar_tiro(arquivo_direto, segmentos_direto)
dados_reverso = processar_tiro(arquivo_reverso, segmentos_reverso)

if dados_direto and dados_reverso:
    print("\n--- Gerando gráfico combinado das dromocronas ---")
    fig1, ax1 = plt.subplots(figsize=(14, 8))
    ax1.scatter(dados_direto['posicoes'], dados_direto['tempos'], label=f'Dados Direto ({arquivo_direto})', marker='o', c='black')
    for nome, ajuste in dados_direto['ajustes'].items():
        ax1.plot(ajuste['x'], ajuste['y'], '-', color=ajuste['cor'], lw=2.5, label=f'Direto - {ajuste["label"]}')
    ax1.scatter(dados_reverso['posicoes'], dados_reverso['tempos'], label=f'Dados Reverso ({arquivo_reverso})', marker='x', c='black')
    for nome, ajuste in dados_reverso['ajustes'].items():
        ax1.plot(ajuste['x'], ajuste['y'], '--', color=ajuste['cor'], lw=2.5, label=f'Reverso - {ajuste["label"]}')
    ax1.set_xlabel('Posição do Receptor (m)')
    ax1.set_ylabel('Tempo (s)')
    ax1.set_title('Análise Combinada das Dromocronas')
    ax1.legend()
    ax1.grid(True)
    ax1.invert_yaxis()
    plt.savefig('dromocronas_combinadas.png', dpi=300, bbox_inches='tight')
    plt.show()

if dados_direto and dados_reverso:
    print("\n--- Unificando resultados para gerar os Modelos Geológicos ---")
    pos_direto, t_direto, vel_direto, int_direto = dados_direto['posicoes'], dados_direto['tempos'], dados_direto['velocidades'], dados_direto['interceptos']
    pos_reverso, t_reverso, vel_reverso, int_reverso = dados_reverso['posicoes'], dados_reverso['tempos'], dados_reverso['velocidades'], dados_reverso['interceptos']
    v1_direta = vel_direto.get("Onda Direta", 0)
    v1_reversa = vel_reverso.get("Onda Direta", 0)
    v1 = (v1_direta + v1_reversa) / 2
    print(f"\n==> Velocidade V1 Média utilizada: {v1:.2f} m/s")
    vd1 = vel_direto.get("Refratada 1", 0)
    vu1 = vel_reverso.get("Refratada 1", 0)
    tid1 = int_direto.get("Refratada 1", 0)
    mascara_refratada_rev1 = (pos_reverso >= segmentos_reverso['Refratada 1']['dist_min']) & (pos_reverso <= segmentos_reverso['Refratada 1']['dist_max'])
    pos_refratada_rev1 = pos_reverso[mascara_refratada_rev1]
    t_refratada_rev1 = t_reverso[mascara_refratada_rev1]
    tiu1 = 0
    if len(pos_refratada_rev1) >= 2:
        _, tiu1 = np.polyfit(distancia_total_AB - pos_refratada_rev1, t_refratada_rev1, 1)
    tempo_reciproco_AB = np.interp(distancia_total_AB, pos_direto, t_direto)
    print(f"Tempo Recíproco (t_AB) estimado: {tempo_reciproco_AB:.4f} s")
    print("\n--- Resultados da 1ª Interface ---")
    if v1 > 0 and vd1 > 0 and vu1 > 0:
        with np.errstate(invalid='ignore'):
            ic1_rad = 0.5 * (np.arcsin(v1/vd1) + np.arcsin(v1/vu1))
            gama1_rad = 0.5 * (np.arcsin(v1/vd1) - np.arcsin(v1/vu1))
        if np.isnan(ic1_rad):
            print("ERRO: Velocidades inconsistentes.")
        else:
            v2_intersecao = v1 / np.sin(ic1_rad)
            h_A1_perp = (tid1 * v1 * np.cos(ic1_rad)) / 2
            h_B1_perp = (tiu1 * v1 * np.cos(ic1_rad)) / 2
            Z_A1 = h_A1_perp / np.cos(gama1_rad)
            Z_B1 = h_B1_perp / np.cos(gama1_rad)
            t_reverso_interp = np.interp(pos_direto, pos_reverso, t_reverso)
            t_v_menos1 = 0.5 * (t_direto - t_reverso_interp + tempo_reciproco_AB)
            incl_v2_real, _ = np.polyfit(pos_direto, t_v_menos1, 1)
            v2_real = 1 / incl_v2_real
            delay_time1 = 0.5 * (t_direto + t_reverso_interp - tempo_reciproco_AB)
            ic1_real_rad = np.arcsin(v1 / v2_real)
            h1x_perp = (delay_time1 * v1) / np.cos(ic1_real_rad)
            Z1x_vertical = h1x_perp / np.cos(gama1_rad)

            print("\n--- Gerando gráfico do Método de Interseção ---")
            plt.figure(figsize=(14, 8))
            ax2 = plt.gca()
            ax2.plot([0, distancia_total_AB], [-Z_A1, -Z_B1], 'o--', color='darkviolet', lw=2, label='Interface 1 (M. Interseção)')
            ax2.set_title('Modelo Geológico - Método de Interseção')
            ax2.set_xlabel('Posição (m)')
            ax2.set_ylabel('Profundidade (m)')
            ax2.legend()
            ax2.grid(True)
            ax2.set_aspect('equal', adjustable='box')
            plt.savefig('interface_intersecao.png', dpi=300, bbox_inches='tight')
            plt.show()

            print("\n--- Gerando gráfico do Método Recíproco ---")
            mascara_corte = (pos_direto >= corte_reciproco_min_m) & (pos_direto <= corte_reciproco_max_m)
            pos_direto_cortado = pos_direto[mascara_corte]
            Z1x_vertical_cortado = Z1x_vertical[mascara_corte]
            
            plt.figure(figsize=(14, 8))
            ax3 = plt.gca()
            
            ax3.plot(pos_direto_cortado, -Z1x_vertical_cortado, 'o-', label='Interface 1 (M. Recíproco)', color='saddlebrown', mfc='tan', lw=2.5, zorder=10)
            
            ax3.set_title('Modelo Geológico - Método Recíproco (Detalhe)')
            ax3.set_xlabel('Posição (m)')
            ax3.set_ylabel('Profundidade (m)')
            
            if eixo_x_min is not None and eixo_x_max is not None:
                ax3.set_xlim(eixo_x_min, eixo_x_max)
            if profundidade_min_plot is not None and profundidade_max_plot is not None:
                ax3.set_ylim(profundidade_min_plot, profundidade_max_plot)
            
            ax3.legend()
            ax3.grid(True)
            plt.savefig('interface_reciproco_detalhe.png', dpi=300, bbox_inches='tight')
            plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

arquivo_direto = 'picks/n13'
segmentos_direto = {
    "Onda Direta": { "dist_min": 0, "dist_max": 20, "cor": "red" },
    "Refratada 1": { "dist_min": 83, "dist_max": 999999, "cor": "darkred" },
}

arquivo_reverso = 'picks/n4'
segmentos_reverso = {
    "Onda Direta": { "dist_min": 83, "dist_max": 99999, "cor": "blue" },
    "Refratada 1": { "dist_min": 0, "dist_max": 70, "cor": "darkblue" },
}

distancia_total_AB = 100.0

corte_reciproco_min_m = 19.0
corte_reciproco_max_m = 85.0

profundidade_min_plot = -20.0
profundidade_max_plot = 5.0

def processar_tiro(nome_do_arquivo, segmentos_para_analise):
    print(f"\n--- Processando o arquivo: {nome_do_arquivo} ---")
    try:
        tempos_brutos, posicoes_brutas = np.loadtxt(nome_do_arquivo, unpack=True)
    except Exception as e:
        print(f"ERRO ao carregar o arquivo: {e}")
        return None
    indices_sorted = np.argsort(posicoes_brutas)
    posicoes = posicoes_brutas[indices_sorted]
    tempos = tempos_brutos[indices_sorted]
    print(f"Arquivo carregado com {len(posicoes)} pontos.")
    resultados = {'posicoes': posicoes, 'tempos': tempos, 'velocidades': {}, 'interceptos': {}}
    for nome, limites in segmentos_para_analise.items():
        mascara = (posicoes >= limites['dist_min']) & (posicoes <= limites['dist_max'])
        pos_segmento, t_segmento = posicoes[mascara], tempos[mascara]
        if len(pos_segmento) < 2:
            print(f"AVISO: Segmento '{nome}' tem pontos insuficientes.")
            continue
        inclinacao, intercepto = np.polyfit(pos_segmento, t_segmento, 1)
        velocidade = abs(1 / inclinacao)
        resultados['velocidades'][nome] = velocidade
        resultados['interceptos'][nome] = intercepto
        print(f"  -> Segmento '{nome}': Velocidade = {velocidade:.2f} m/s")
    return resultados

dados_direto = processar_tiro(arquivo_direto, segmentos_direto)
dados_reverso = processar_tiro(arquivo_reverso, segmentos_reverso)

if dados_direto and dados_reverso:
    print("\n--- Unificando resultados para gerar o Modelo Geológico ---")
    pos_direto, t_direto, vel_direto, int_direto = dados_direto['posicoes'], dados_direto['tempos'], dados_direto['velocidades'], dados_direto['interceptos']
    pos_reverso, t_reverso, vel_reverso, int_reverso = dados_reverso['posicoes'], dados_reverso['tempos'], dados_reverso['velocidades'], dados_reverso['interceptos']
    v1_direta = vel_direto.get("Onda Direta", 0)
    v1_reversa = vel_reverso.get("Onda Direta", 0)
    v1 = (v1_direta + v1_reversa) / 2
    print(f"\n==> Velocidade V1 Média utilizada: {v1:.2f} m/s")
    vd1 = vel_direto.get("Refratada 1", 0)
    vu1 = vel_reverso.get("Refratada 1", 0)
    tid1 = int_direto.get("Refratada 1", 0)
    mascara_refratada_rev1 = (pos_reverso >= segmentos_reverso['Refratada 1']['dist_min']) & (pos_reverso <= segmentos_reverso['Refratada 1']['dist_max'])
    pos_refratada_rev1 = pos_reverso[mascara_refratada_rev1]
    t_refratada_rev1 = t_reverso[mascara_refratada_rev1]
    tiu1 = 0
    if len(pos_refratada_rev1) >= 2:
        _, tiu1 = np.polyfit(distancia_total_AB - pos_refratada_rev1, t_refratada_rev1, 1)
    tempo_reciproco_AB = np.interp(distancia_total_AB, pos_direto, t_direto)
    print(f"Tempo Recíproco (t_AB) estimado: {tempo_reciproco_AB:.4f} s")
    print("\n--- Resultados da 1ª Interface ---")
    if v1 > 0 and vd1 > 0 and vu1 > 0:
        with np.errstate(invalid='ignore'):
            ic1_rad = 0.5 * (np.arcsin(v1/vd1) + np.arcsin(v1/vu1))
            gama1_rad = 0.5 * (np.arcsin(v1/vd1) - np.arcsin(v1/vu1))
        if np.isnan(ic1_rad):
            print("ERRO: Velocidades inconsistentes.")
        else:
            v2_intersecao = v1 / np.sin(ic1_rad)
            h_A1_perp = (tid1 * v1 * np.cos(ic1_rad)) / 2
            h_B1_perp = (tiu1 * v1 * np.cos(ic1_rad)) / 2
            Z_A1 = h_A1_perp / np.cos(gama1_rad)
            Z_B1 = h_B1_perp / np.cos(gama1_rad)
            t_reverso_interp = np.interp(pos_direto, pos_reverso, t_reverso)
            t_v_menos1 = 0.5 * (t_direto - t_reverso_interp + tempo_reciproco_AB)
            incl_v2_real, _ = np.polyfit(pos_direto, t_v_menos1, 1)
            v2_real = 1 / incl_v2_real
            delay_time1 = 0.5 * (t_direto + t_reverso_interp - tempo_reciproco_AB)
            ic1_real_rad = np.arcsin(v1 / v2_real)
            h1x_perp = (delay_time1 * v1) / np.cos(ic1_real_rad)
            Z1x_vertical = h1x_perp / np.cos(gama1_rad)

            print("\n--- Gerando gráfico final comparativo ---")
            
            mascara_corte = (pos_direto >= corte_reciproco_min_m) & (pos_direto <= corte_reciproco_max_m)
            pos_direto_cortado = pos_direto[mascara_corte]
            Z1x_vertical_cortado = Z1x_vertical[mascara_corte]
            
            plt.figure(figsize=(14, 8))
            ax = plt.gca()
            
            ax.plot(pos_direto_cortado, -Z1x_vertical_cortado, 'o-', label='Interface 1 (M. Recíproco)', color='saddlebrown', mfc='tan', lw=2.5, zorder=10)
            ax.plot([0, distancia_total_AB], [-Z_A1, -Z_B1], '--', color='darkviolet', lw=2, label='Interface 1 (M. Interseção)')
            
            texto_reciproco = (f"M. Recíproco:\nV1={v1:.0f} V2={v2_real:.0f}\nProf. Mín={np.min(Z1x_vertical_cortado):.1f}m Prof. Máx={np.max(Z1x_vertical_cortado):.1f}m")
            texto_intersecao = (f"M. Interseção:\nV2(est)={v2_intersecao:.0f} γ₁={np.degrees(gama1_rad):.1f}°\nZₐ₁={Z_A1:.1f}m Zᵦ₁={Z_B1:.1f}m")

            ax.set_title('Figura Final: Comparação dos Modelos Geológicos')
            ax.text(0.98, 0.98, texto_reciproco, transform=ax.transAxes, fontsize=10, verticalalignment='top', horizontalalignment='right', bbox=dict(boxstyle='round,pad=0.5', fc='skyblue', alpha=0.8))
            ax.text(0.02, 0.98, texto_intersecao, transform=ax.transAxes, fontsize=10, verticalalignment='top', horizontalalignment='left', bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.8))
            
            ax.set_xlabel('Posição (m)')
            ax.set_ylabel('Profundidade (m)')
            ax.legend(loc='lower center')
            ax.grid(True)
            
            if profundidade_min_plot is not None and profundidade_max_plot is not None:
                ax.set_ylim(profundidade_min_plot, profundidade_max_plot)
            
            ax.set_aspect('equal', adjustable='box')

            plt.savefig('interfaces_finais_comparativo.png', dpi=300, bbox_inches='tight')
            plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

analises = {
    'n12': {
        'arquivo': 'picks/nu=12.txt',
        'cor_dados': 'blue',
        'segmentos': {
            'Onda Direta': {'dist_min': 86, 'dist_max': 99999, 'cor': 'darkblue'},
            'Refratada 1': {'dist_min': 74, 'dist_max': 86, 'cor': 'blue'},
            'Refratada 2': {'dist_min': 10, 'dist_max': 74, 'cor': 'lightblue'}
        }
    },
    
    'n13': {
        'arquivo': 'picks/n13',
        'cor_dados': 'red',
        'segmentos': {
            'Onda Direta': {'dist_min': 0, 'dist_max': 0, 'cor': 'darkred'},
            'Refratada 1': {'dist_min': 21, 'dist_max': 65, 'cor': 'red'},
            'Refratada 2': {'dist_min': 80, 'dist_max': 99999, 'cor': 'lightsalmon'}
        },
    }
}

plt.figure(figsize=(14, 8))
ax = plt.gca()
print("--- Iniciando Análise Comparativa de Tiros ---")

for nome_tiro, config in analises.items():
    nome_arquivo = config['arquivo']
    segmentos_para_analise = config['segmentos']
    cor_dados = config['cor_dados']
    
    print(f"\n--- Analisando o '{nome_tiro}': {nome_arquivo} ---")
    
    try:
        tempos_brutos, posicoes_brutas = np.loadtxt(nome_arquivo, unpack=True)
    except Exception as e:
        print(f"ERRO ao carregar o arquivo '{nome_arquivo}': {e}")
        continue

    indices_sorted = np.argsort(posicoes_brutas)
    posicoes = posicoes_brutas[indices_sorted]
    tempos = tempos_brutos[indices_sorted]
    
    ax.scatter(posicoes, tempos, label=f'Dados {nome_tiro}', marker='o', alpha=0.7, c=cor_dados)

    for nome_segmento, limites in segmentos_para_analise.items():
        dist_min = limites['dist_min']
        dist_max = limites['dist_max']
        cor_reta = limites['cor']
        
        mascara = (posicoes >= dist_min) & (posicoes <= dist_max)
        pos_segmento = posicoes[mascara]
        t_segmento = tempos[mascara]
        
        if len(pos_segmento) < 2:
            print(f"AVISO: Segmento '{nome_segmento}' do '{nome_tiro}' tem pontos insuficientes.")
            continue
            
        inclinacao, intercepto = np.polyfit(pos_segmento, t_segmento, 1)
        velocidade = abs(1 / inclinacao)
        
        print(f"  -> Segmento '{nome_segmento}': Velocidade = {velocidade:.2f} m/s")
        
        ax.plot(pos_segmento, inclinacao * pos_segmento + intercepto, '-', color=cor_reta, lw=2.5, 
                label=f'Ajuste {nome_tiro} - {nome_segmento} ({velocidade:.0f} m/s)')

ax.set_xlabel('Posição do Receptor (m)')
ax.set_ylabel('Tempo (s)')
ax.set_title('Análise Comparativa de Dromocronas')
ax.legend()
ax.grid(True)
ax.invert_yaxis()

nome_figura = 'analise_comparativa.png'
plt.savefig(nome_figura, dpi=300, bbox_inches='tight')
print(f"\nGráfico comparativo salvo como: {nome_figura}")

plt.show()
```
