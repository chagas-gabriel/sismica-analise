```python
# ==============================================================================
# CÉLULA 1: PICKING INTERATIVO COM PRÉ-PROCESSAMENTO
# ==============================================================================

%matplotlib widget

import obspy
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

# ==============================================================================
# PAINEL DE CONTROLE AVANÇADO
# ==============================================================================

# --- 1. CONFIGURAÇÕES BÁSICAS DO TIRO ---
nome_do_arquivo_su = 'plots-files/2.su'
posicao_da_fonte = 49.0

# --- 2. FILTRO PASSA-BANDA (OPCIONAL) ---
# Mude para True para ativar o filtro de frequência
aplicar_filtro_banda = True
# Frequências de corte do filtro (f1, f2, f3, f4) como no SU.
# A banda passante principal é entre f2 e f3.
f1, f2, f3, f4 = 20, 30, 130, 150

# --- 3. SOBREPOR PICKS (OPCIONAL) ---
# Mude para True para carregar e plotar um arquivo de picks existente
sobrepor_picks_antigos = False
# Coloque o nome do arquivo de picks que você quer sobrepor
arquivo_picks_antigos = 'picks_10su.txt' 

# --- 4. CONFIGURAÇÕES DE VISUALIZAÇÃO E ARRANJO ---
ganho_visual = 30

NG = 96
dx = 1.0
g1 = 0.0

# ==============================================================================
# CARREGAMENTO E PROCESSAMENTO
# ==============================================================================
print(f"--- Carregando o arquivo: {nome_do_arquivo_su} ---")
stream = obspy.read(nome_do_arquivo_su, format='SU')
print("Arquivo SU carregado com sucesso!")

# --- APLICAÇÃO DO FILTRO (SE ATIVADO) ---
if aplicar_filtro_banda:
    try:
        # ObsPy usa um filtro Butterworth. Usamos f2 e f3 como as frequências da banda principal.
        stream.filter('bandpass', freqmin=f2, freqmax=f3, corners=4, zerophase=True)
        print(f"Filtro Passa-Banda aplicado entre {f2} Hz e {f3} Hz.")
    except Exception as e:
        print(f"ERRO ao aplicar o filtro: {e}")

# --- ATRIBUIÇÃO DE GEOMETRIA ---
pos_geofones = g1 + np.arange(NG) * dx
print("\nAtribuindo geometria aos traços...")
for i, tr in enumerate(stream):
    pos_geofone = pos_geofones[i]
    tr.stats.offset = abs(pos_geofone - posicao_da_fonte)
    tr.stats.geophone_position = pos_geofone
stream.sort(keys=['geophone_position'])
print("Geometria atribuída e traços ordenados por POSIÇÃO DO GEOFONE.")

# ==============================================================================
# CRIAÇÃO DA FIGURA E DO PICKING INTERATIVO
# ==============================================================================
print("\n>>> GRÁFICO INTERATIVO ABAIXO <<<")

picks = [] 
pick_plots = []
fig, ax = plt.subplots(figsize=(10, 7))
max_val = max(abs(tr.data).max() for tr in stream) if len(stream) > 0 else 1
all_geophone_positions = np.array([tr.stats.geophone_position for tr in stream])

for i, tr in enumerate(stream):
    geophone_pos = tr.stats.geophone_position
    ax.plot(geophone_pos + (tr.data / max_val) * (dx * ganho_visual), tr.times(), 'k-', linewidth=0.8, label='_nolegend_')

# --- SOBREPOSIÇÃO DOS PICKS ANTIGOS (SE ATIVADO) ---
if sobrepor_picks_antigos:
    try:
        tempos_antigos, posicoes_antigas = np.loadtxt(arquivo_picks_antigos, unpack=True)
        ax.plot(posicoes_antigas, tempos_antigos, '+', color='cyan', markersize=10, mew=2, label=f'Picks Antigos ({arquivo_picks_antigos})')
        print(f"Arquivo de picks '{arquivo_picks_antigos}' sobreposto no gráfico.")
    except Exception as e:
        print(f"AVISO: Não foi possível carregar ou plotar o arquivo de picks antigos: {e}")

# --- FUNÇÃO DE CALLBACK PARA O TECLADO ---
def on_key(event):
    if event.key == 'a':
        ix, iy = event.xdata, event.ydata
        if ix is not None and iy is not None:
            idx_mais_proximo = np.argmin(np.abs(all_geophone_positions - ix))
            posicao_correta = all_geophone_positions[idx_mais_proximo]
            picks.append({'pos': posicao_correta, 't': iy})
            plot_obj = ax.plot(posicao_correta, iy, 'rx', markersize=8, markeredgewidth=2, label='_nolegend_')
            pick_plots.append(plot_obj[0])
            print(f"Pick ADICIONADO: Posição={posicao_correta:.2f} m, Tempo={iy:.4f} s")
            plt.gcf().canvas.draw_idle()
    elif event.key == 'd':
        if picks:
            picks.pop(); last_plot = pick_plots.pop(); last_plot.remove()
            print("Último pick REMOVIDO.")
            plt.gcf().canvas.draw_idle()

fig.canvas.mpl_connect('key_press_event', on_key)
ax.set_title(f"Picking Interativo - Arquivo: {nome_do_arquivo_su} (Fonte em {posicao_da_fonte}m)")
ax.set_xlabel("Posição do Geofone (m)")
ax.set_ylabel("Tempo (s)")
ax.invert_yaxis()
ax.set_xlim(g1 - dx, g1 + (NG-1)*dx + dx)
ax.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
